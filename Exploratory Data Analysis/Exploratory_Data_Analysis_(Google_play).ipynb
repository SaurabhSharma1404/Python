# **Exploratory Data Analysis**

**EDA** is the process of looking at your data carefully to understand what’s inside, spot patterns, and find problems before doing deeper analysis.

EDA is like detective work for data – you explore, clean, and understand your dataset before using it to make decisions or predictions.

**Types are-:**

1. **Univariate Analysis** -: Looking at ONE variable (column) at a time.

**Goal:** Understand its distribution (how values are spread).

**Used for:** Spotting outliers, missing values, and basic patterns.

**Examples of techniques:**

* For numbers: Histograms, boxplots, summary stats (mean, median, min, max).

* For categories: Bar charts, frequency tables.

**Example:** Looking at “Age” column alone – finding the average age, how many people are young vs old.

2. **Bivariate Analysis**-: Looking at TWO variables together.

**Goal:** See relationships or comparisons between two columns.

**Used for**: Finding patterns (e.g., does income depend on education level?).

**Examples of techniques:**

* For number vs number: Scatter plots, correlation.

* For category vs number: Boxplots, bar plots.

* For category vs category: Crosstabs, grouped bar charts.

**Example:** Comparing “Age” and “Salary” to see if older people earn more.

3. **Multivariate Analysis**-: Looking at MORE THAN TWO variables at once.

**Goal**: Understand complex relationships between many columns.

**Used for:** Detecting deeper patterns, interactions, or clusters.

Examples of techniques:

* Heatmaps (show correlation between many numeric variables)

* Pairplots (scatter plots for multiple numeric columns)

* Advanced methods like PCA (Principal Component Analysis) for dimensionality reduction

```python
# we write a fucntion and we apply this function on the size column to remove the Mb and kb and covert all in kb

def size_app(item):
  if item[-1] == 'M':
    result = float(str(item).replace('M',""))
    result  = result*1024
    return result
  elif item[-1] == 'k':
    result = float(str(item).replace('k',""))
    return result
  else:
    return str(np.nan)
```
