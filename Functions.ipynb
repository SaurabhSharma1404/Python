Function-:

* A function is a block of code that does something.

* It is not tied to any object.

* You define it using def outside of any class.


```python
# example of a fuction
def greet(name): # fuction is not defined inside an object
    return "Hello " + name

print(greet("Saurabh Sharma"))
```

Method-:

* A method is a function inside a class.

* It works with an object
* A method definition always includes self as its first parameter

* Method is called by its name, but it is associated to an object

```python
# example of a method
class Person:
    def greet(self, name): # this is called as a method because it is inside a class
        return "Hello " + name

p = Person()
print(p.greet("Saurabh Sharma"))
```

# **Answer 6**

**Generators-:** They are type of iterator that can produce a sequence of values one at a time, rather than storing all in a memory at once.

* General fuction uses a return statement but a generator function uses **yield**

```python
# defining a generator
def squ(n):
  for i in range(n):
    yield i**2  # generator

gen = squ(5)
print(next(gen))
print(next(gen)) # output will be printed one by one using the NEXT
print(next(gen))
print(next(gen))
print(next(gen))
```

# **Answer 7**

Advantages of using generators over regular functions are-:

* **Memory usage-:** Regular function gives output in one go whcih lead to more memory usage but on other side generators provide the output one by one which will take less memeory comapare to regular functions.

* **Execution time-:** Generators take very less execution time when there are lot of computations but the regular function take a lot of execution time when there is lot of computaions.

* **Pause and Resume-:** Generators pause at each yield and resume where they left off.



```python
# lets take an example
def infinite_numbers():
    i = 0
    while True:
        yield i
        i += 1

gen = infinite_numbers()
print(next(gen))
print(next(gen))
print(next(gen))

# if you try this using a regular function then it will crash your memory
```

# **Answer 9**    

**Purpose of map():**

The main purpose of map() is to apply a given function to all items in an input iterable (like a list) and return a new iterable (a map object) containing the results. It saves you from writing explicit loops to process each item individually, making your code more concise and often more readable.

**Usage of map():**

The map() function takes two main arguments:

**A function:** This is the function that will be applied to each item in the iterable. This function can be a regular function you've defined with def, or it can be an anonymous function created using lambda.

**An iterable:** This is the sequence (like a list, tuple, or string) whose items you want to process.

```python
# example of map function
def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(list(squared_numbers))

# without using a loop we can iterate through with the help of map functions
```

```python
def prime(n):
    if n < 2: # 0 and 1 are nor prime numbers
        return "not prime"
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return "not prime"
    return "prime"

for num in range(1, 201):
    print(num, prime(num))
```

# **Answer 5**

```python
class Fibonacci:
    def __init__(self, total_terms):
        self.total_terms = total_terms
        self.current_term = 0
        self.first = 0
        self.second = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.total_terms:
            raise StopIteration
        if self.current_term == 0:
            self.current_term += 1
            return self.first
        elif self.current_term == 1:
            self.current_term += 1
            return self.second
        else:
            next_number = self.first + self.second
            self.first = self.second
            self.second = next_number
            self.current_term += 1
            return next_number

# Asking the user for number of terms
n = int(input("How many Fibonacci numbers do you want? "))

# Creating the iterator
fib_seq = Fibonacci(n)

# Printing the sequence
print("Fibonacci sequence:")
for num in fib_seq:
    print(num)
```

# **Answer 7**

```python
def read_file_line_by_line(filename):
    try:
        with open(filename, 'r') as file:
            for line in file:
                yield line.strip()  # remove newline characters
    except FileNotFoundError:
        print("File not found.")

file_name = input("Enter the name of the file to read: ")

for line in read_file_line_by_line(file_name):
    print(line)
```

```python
def vowel(char):
    vowels = 'aeiouAEIOU'
    return char not in vowels

text = input("Enter a string: ")

# Use filter to remove vowels
result = ''.join(filter(vowel, text))

print("String without vowels:", result)
```
